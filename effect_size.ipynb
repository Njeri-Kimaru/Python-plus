```python

import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
%matplotlib inline
np.random.seed(10)
```

```python
male_mean = 178
male_std = 7.8
```

```python
female_mean = 163
female_std = 7.3
```

```python
male_height = stats.norm(male_mean,male_std)
female_height = stats.norm(female_mean,female_std)
```

```python
def evaluate_pdf(rv, x=4):
    #evaluating the mean and std of the random variables.
    mean = rv.mean() 
    sd = rv.std()
    #coming up with a normal distribution using numpy to calculate evenly spaced numbers over the specified interval
    xs = np.linspace(mean - x*sd, mean + x*sd)
    #calculate the peak of the normal distribution 
    yx = rv.pdf(xs)
    return (xs, yx)
    
```

```python
#Male height
xs, yx = evaluate_pdf(male_height)
plt.plot(xs, yx, label='male', linewidth=4, color="#5d06d6") 
 
#Female height 
xs, yx = evaluate_pdf(female_height)
plt.plot(xs, yx, label='female', linewidth=4, color="#f44444")
 
plt.xlabel('height (cm)')
```

```python
mean2,sd2 = female_sample.mean(),female_sample.std()
mean2,sd2
```
